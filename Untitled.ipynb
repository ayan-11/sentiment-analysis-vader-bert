```python
import pandas as pd

# Load the data
data = pd.read_excel('Final_output.xlsx', sheet_name='Sheet1')

# Function to categorize sentiment based on polarity score
def categorize_sentiment(polarity_score):
    if polarity_score > 0.6:
        return 'Very Positive'
    elif 0.2 < polarity_score <= 0.6:
        return 'Positive'
    elif 0 < polarity_score <= 0.2:
        return 'Slightly Positive'
    elif polarity_score == 0:
        return 'Neutral'
    elif -0.2 < polarity_score < 0:
        return 'Slightly Negative'
    else:
        return 'Negative'

# Apply the function to the Polarity Score column
data['Sentiment'] = data['Polarity Score'].apply(categorize_sentiment)

# Display the first few rows with the new Sentiment column
data.head()
```

```python
import matplotlib.pyplot as plt

# Count the number of each sentiment
sentiment_counts = data['Sentiment'].value_counts()

# Plot the sentiment distribution
plt.figure(figsize=(10, 6))
sentiment_counts.plot(kind='bar', color='skyblue')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sentiment_counts.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'yellow', 'orange', 'red', 'lightcoral'])
plt.title('Sentiment Distribution')
plt.ylabel('')
plt.show()
```
