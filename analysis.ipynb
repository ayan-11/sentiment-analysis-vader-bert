```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Final_output.xlsx', sheet_name='Sheet1')
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
axs[0, 0].hist(data['Positive Score'], bins=20, color='green', edgecolor='black')
axs[0, 0].set_title('Positive Score Distribution')
axs[0, 0].set_xlabel('Positive Score')
axs[0, 0].set_ylabel('Frequency')
axs[0, 1].hist(data['Negative Score'], bins=20, color='red', edgecolor='black')
axs[0, 1].set_title('Negative Score Distribution')
axs[0, 1].set_xlabel('Negative Score')
axs[0, 1].set_ylabel('Frequency')
axs[1, 0].hist(data['Polarity Score'], bins=20, color='blue', edgecolor='black')
axs[1, 0].set_title('Polarity Score Distribution')
axs[1, 0].set_xlabel('Polarity Score')
axs[1, 0].set_ylabel('Frequency')
axs[1, 1].hist(data['Subjectivity Score'], bins=20, color='purple', edgecolor='black')
axs[1, 1].set_title('Subjectivity Score Distribution')
axs[1, 1].set_xlabel('Subjectivity Score')
axs[1, 1].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Positive Score'], bins=20, color='green', edgecolor='black')
plt.title('Positive Score Distribution')
plt.xlabel('Positive Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram of Negative Scores
plt.figure(figsize=(10, 6))
plt.hist(data['Negative Score'], bins=20, color='red', edgecolor='black')
plt.title('Negative Score Distribution')
plt.xlabel('Negative Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram of Polarity Scores
plt.figure(figsize=(10, 6))
plt.hist(data['Polarity Score'], bins=20, color='blue', edgecolor='black')
plt.title('Polarity Score Distribution')
plt.xlabel('Polarity Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram of Subjectivity Scores
plt.figure(figsize=(10, 6))
plt.hist(data['Subjectivity Score'], bins=20, color='purple', edgecolor='black')
plt.title('Subjectivity Score Distribution')
plt.xlabel('Subjectivity Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter Plot of Positive vs Negative Scores
plt.figure(figsize=(10, 6))
plt.scatter(data['Positive Score'], data['Negative Score'], color='blue')
plt.title('Positive vs Negative Scores')
plt.xlabel('Positive Score')
plt.ylabel('Negative Score')
plt.show()
```

```python
# Scatter Plot of Polarity Score vs Subjectivity Score
plt.figure(figsize=(10, 6))
plt.scatter(data['Polarity Score'], data['Subjectivity Score'], color='orange')
plt.title('Polarity Score vs Subjectivity Score')
plt.xlabel('Polarity Score')
plt.ylabel('Subjectivity Score')
plt.show()
```

```python
# Bar Chart of Average Sentence Length
plt.figure(figsize=(10, 6))
data['Average Sentence Length'].plot(kind='bar', color='cyan')
plt.title('Average Sentence Length for each URL')
plt.xlabel('URL ID')
plt.ylabel('Average Sentence Length')
plt.xticks(rotation=90)
plt.show()
```

```python
# Bar Chart of Word Count
plt.figure(figsize=(10, 6))
data['Word Count'].plot(kind='bar', color='magenta')
plt.title('Word Count for each URL')
plt.xlabel('URL ID')
plt.ylabel('Word Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Pie Chart of Positive and Negative Scores
positive_sum = data['Positive Score'].sum()
negative_sum = data['Negative Score'].sum()

plt.figure(figsize=(8, 8))
plt.pie([positive_sum, negative_sum], labels=['Positive Score', 'Negative Score'], autopct='%1.1f%%', colors=['green', 'red'])
plt.title('Proportion of Positive and Negative Scores')
plt.show()
```

```python
# Box Plot of Polarity Scores
plt.figure(figsize=(10, 6))
plt.boxplot(data['Polarity Score'])
plt.title('Box Plot of Polarity Scores')
plt.ylabel('Polarity Score')
plt.show()
```
